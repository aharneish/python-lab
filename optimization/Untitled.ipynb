```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import minimize
from mpl_toolkits import mplot3d
```

# problem 1
$$
f(x)=4x^4-x^2+5
$$

```python
f=lambda x:4*x**4-x**2+5
x=np.linspace(-5,5,100)
F=f(x)
plt.plot(x,F)
```

```python
# getting the minima
x0=np.random.random(1)
res=minimize(f,x0)
res
```

```python
# getting the maxima
x0=np.random.random(1)
res=minimize(f,-x0)
res
```

# problem 2
$$
f(x)=x_1^3+x_2^3-3x_1x_2
$$

```python
f=lambda x,y: x**3+y**3-3*x*y #taking the values of x1 as x and x2 as y
x=np.linspace(-10,10,100)
y=x.copy()
F=f(x,y)
fig= plt.figure(figsize=(10,10))
ax = plt.axes(projection='3d')
ax.grid()
surf = ax.plot_surface(x,y,F, cmap= plt.cm.rainbow)

# set axes label
ax.set_xlabel('x', labelpad=20)
ax.set_ylabel('y', labelpad=20)
ax.set_zlabel('z', labelpad=20)

fig.colorbar(surf, shrink=0.5, aspect=8) # adding colorbar

plt.show()
```
