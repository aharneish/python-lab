# Binomial Binom(n,p)

```python
def plot_binomial(n,p):
    """
    this function will calculate the binomial distribution for the given n number of trials the probability of success is p
    n is a integer value
    p is a float value
    Author: Aharneish Abburu
    Date: 8/9/2022
    """
    x=[]
    for i in range(0,n+1):
        x.append(i)
    y=[]
    for i in range(len(x)):
        y.append(comb(n,x[i])*(p**x[i])*(1-p)**(n-x[i]))
    plt.bar(x,y,width=0.2,color='red')
    plt.plot(x,y)
    plt.title(f"binomial distribution for n={n} and p={p}")
    plt.xlabel("x")
    plt.ylabel("the probability")
    plt.show()
```

```python
n=int(input("enter the number of times the binomial trail has taken place: "))
p=float(input("enter the probability of success of the trails: "))
plot_binomial(n,p)
```

# Normal Distribution

```python
#normal distribution
# generate 100 linspace points between (mu-4sigma,mu+4sigma)
#define the function f(x)
#plot x,f(x)
def plot_normal(m,s):
    """
    this function will calculate the normal distribution for the given mean m and standard deviation s
    m is a float value
    s is a float value
    Author: Aharneish Abburu
    Date: 8/9/2022
    """
    y=[]
    X=np.linspace(m-4*s,m+4*s,100)
    f=lambda x:(1/(s*sqrt(2*pi))*exp(-(1/2)*((x-m)/s)**2))
    for i in range(len(X)):
        y.append(f(X[i]))
    plt.plot(X,y)
    plt.title(f"plot for normal distribution with parameters ({m},{s})")
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.show()
```

```python
m=float(input("enter the mean for the normal distribution: "))
s=float(input("enter the standard deviation for the normal distribution: "))
plot_normal(m,s)
```

# Poisson Distribution


```python
def plot_poisson(l):
    """
    this function will calculate the poisson distribution for the given lambda l
    l is a int value
    Author: Aharneish Abburu
    Date: 8/9/2022
    """
    def fact(a):
        """this is a factorial function for getting the fatorial of a given number a
            a is a int value
            Author: Aharneish Abburu
            Date: 8/9/2022
        """
        ans=1
        if a==0:
            return ans
        else:
            while a!=0:
                ans=ans*a
                a=a-1
            return ans
    
    y=[]
    k=int(input("enter the x limits x should be grater than 0:" ))
    X=np.arange(k)
    f=lambda x:exp(-l)*l/fact(x)
    for i in range(len(X)):
        y.append(f(X[i]))
    plt.plot(X,y)
    plt.title(f"plot for poisson with parameters({l})")
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.show()
```

```python
l=int(input("enter the lambda for the poisson distribution: "))
plot_poisson(l)
```

# Exponential Distribution


$$f(x)=\left\{\begin{array}{cl}
\lambda e^{-\lambda x} & x\ge0\\
0 & x<0\\
\end{array}\right.$$

```python
def plot_exponential(l,k):
    """
    this function will calculate the exponential distribution for the given lambda l and k 
    l is a int value
    k is a int value
    Author: Aharneish Abburu
    Date: 8/9/2022
    """
    X=np.arange(0,k)
    y=[]
    f=lambda x,l:l*exp(-l)*exp(x)
    for i in range(len(X)):
        y.append(f(X[i],l))
    plt.plot(X,y)
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.show()
```

```python
l=int(input("enter the lambda value for the exponential distribution: "))
x=int(input("enter the x for the exponential distribution it should be greater than zero: "))
plot_exponential(l,x)
```

# Gamma Distribution
formula:
$${f(x,\alpha,\lambda)={\lambda ^ \alpha  x ^{\alpha - 1}  e ^  {-\lambda x}}/\gamma (\alpha)}$$

```python
def gamma(a):
    """
    this function will calculate the gamma function of aa given alpha value
    alpha is a int value
    Author: Aharneish Abburu
    Date: 11/9/2022
    """
    x=symbols('x')
    g_x=x^(a-1)*math.exp(-x)
    ans=integrate(g_x,(x,0,oo))
    print(f"the gamma value is {ans}")
```

```python
def plot_gamma():
    #x=int(input("enter the x value for gamma: "))
    l=int(input("enter the lambda value for the gamma distribution: "))
    a=float(input("enter the alpha value for the gamma distribution: "))
    X=np.linspace(-100,100,1000)
    y=[]
    f=lambda x: l**a*x**(a-1)*exp(-l)*exp(x)
    for i in range(len(X)):
        y.append(f(X[i]))
    plt.plot(X,y)
    plt.xlabel("X")
    plt.ylabel('F(x,alpha,lambda)')
    plt.show()
```
